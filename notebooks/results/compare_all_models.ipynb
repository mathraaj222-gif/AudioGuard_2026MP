# AudioGuard — Final Results & Model Selection
**Purpose**: Combine leaderboards from both TCA and SER tracks. Select the top-performing model from each for the eventual fusion layer. This is the final step in the local training pipeline.

**Output**: `./outputs/final_leaderboard.csv`

```python
import os, json
import pandas as pd

tca = pd.read_csv("../tca/tca_leaderboard.csv")
ser = pd.read_csv("../ser/ser_leaderboard.csv")
df = pd.concat([tca, ser])
df.to_csv("../../outputs/final_leaderboard.csv", index=False)
print("=== RECOMMENDED MODELS FOR FUSION LAYER ===")
print(f"TCA: {tca.iloc[0]['model_name']} (F1: {tca.iloc[0]['f1_macro']:.4f})")
print(f"SER: {ser.iloc[0]['model_name']} (F1: {ser.iloc[0]['f1_macro']:.4f})")
```
