# SER — Evaluate All Models
## AudioGuard FYP | SER Track
**Purpose**: Aggregate results from S1–S7. Filters `training_summary.json` for the SER track, builds a leaderboard CSV, and visualises performance across different architectures (LSTM vs Transformers).

**Output**: `./outputs/ser_leaderboard.csv`

```python
import os, json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

with open("../../outputs/training_summary.json", "r") as f:
    data = json.load(f)
df = pd.DataFrame([r for r in data if r['track'] == 'SER'])
print(df[['model_id', 'f1_macro', 'accuracy']])
df.to_csv("../../outputs/ser_leaderboard.csv", index=False)
```

```python
df.sort_values("f1_macro", ascending=False).plot(kind="bar", x="model_id", y="f1_macro")
plt.title("SER F1 Macro Leaderboard"); plt.show()
```
